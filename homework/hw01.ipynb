```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# basis functions


def phi(i, j, x, y):
    return np.sin(i * np.pi * x) * np.sin(j * np.pi * y)


def a_i_j(i, j, k, l):
    if i == k and j == l:
        return np.pi**2 / 4 * (i**2 + j**2)
    else:
        return 0


def b(i, j):
    if j % 2 == 1 and i % 2 == 1:
        return 4 / (np.pi**2 * i * j)
    else:
        return 0


def x_i_j(i, j):
    return b(i, j) / a_i_j(i, j, i, j)
```

```python
N = 10

basis_functions_index = [(i, j) for i in range(1, N + 1) for j in range(1, N + 1)]

# basis_functions_index= [(1,1),(1,3),(3,1),(3,3)]
n_func = len(basis_functions_index)


A = np.zeros([n_func, n_func])

for m, row in enumerate(A):
    k, l = basis_functions_index[m]
    for n, item in enumerate(row):
        i, j = basis_functions_index[n]
        A[m][n] = a_i_j(i, j, k, l)

[A[i][i] for i in range(1, n_func)]

x = [x_i_j(i, j) for i, j in basis_functions_index]
print(x)
```

```python
X = np.linspace(0, 1, num=20)
Y = np.linspace(0, 1, num=20)
X, Y = np.meshgrid(X, Y)

Z = 0
for n, (i, j) in enumerate(basis_functions_index):
    Z += x[n] * phi(i, j, X, Y)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z)

pass
```
